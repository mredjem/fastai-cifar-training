```python
data = ImageDataBunch.from_folder(path=path, valid='test', ds_tfms=get_transforms()).normalize(imagenet_stats)
```

```python
data.show_batch(rows=3, figsize=(7, 6))
```

```python
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```
